```python
#eCommerce data analysis
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Data
file_path='eCommerce Store1.xlsx'
df=pd.read_excel(file_path)
```

```python
#preview data
print('First 5 rows\n',df.head())
print('Last 5 rows:\n',df.tail())
print("Summary Statistics:\n",df.describe())
print("\nColumns:\n",df.columns)
```

```python
df['Gender']=df['Gender'].replace({'W':'women','M':'Men'})
print(df)
```

```python
df['Qty']=df['Qty'].replace({'One':1,'Two':2})
```

```python
df['ship-city']=df['ship-city'].str.upper()
df['ship-state']=df['ship-state'].str.upper()
print(df)
```

```python
def categorize_age(age):
    if age>60:
        return 'Above 60'
    elif age>=20:
        return '20-60'
    return 'Below 20'

df['Age Group']=df['Age'].apply(categorize_age)
print(df)
```

```python
df['Date']=pd.to_datetime(df['Date'],format='%d-%m-%Y')
df['Month']=df['Date'].dt.month_name()
print(df)
```

```python
df['Business Type']=df['B2B'].replace({False:'B2C',True:'B2B'})
print(df)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.barplot(data=df,x='Gender',y='Amount',estimator=sum,hue='Gender',palette='pastel',legend=False)
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales(INR)')

plt.subplot(1,2,2)
sns.countplot(data=df,x='Gender',hue='Gender',palette='pastel',legend=False)
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Orders')
plt.tight_layout()
plt.show()
```

```python
#order status distribution
plt.figure(figsize=(8,6))
sns.countplot(data=df,x='Status',hue='Status',palette='viridis',legend=False)
plt.title('Order status distribution')
plt.xlabel('Order Status')
plt.ylabel('Count')
plt.show()
```

```python
#Top 5 states By Sales
top_states=df.groupby('ship-state')['Amount'].sum().nlargest(5).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=top_states,x='Amount',y='ship-state',hue='ship-state',palette='rocket',legend=False)
plt.title('Top 5 States by Sales')
plt.xlabel('Total Sales(INR)')
plt.ylabel('State')
plt.show()
```

```python
#Top 5 states By Sales
top_cities=df.groupby('ship-city')['Amount'].sum().nlargest(5).reset_index()
plt.figure(figsize=(10,6))
sns.barplot(data=top_cities,x='Amount',y='ship-city',hue='ship-city',palette='magma',legend=False)
plt.title('Top 5 Cities by Sales')
plt.xlabel('Total Sales(INR)')
plt.ylabel('State')
plt.show()
```

```python
#Relationship:Age group vs sales(Better Visualization)
plt.figure(figsize=(10,6))
sns.barplot(data=df,x='Age Group',y='Amount',estimator=sum,hue='Age Group',palette='Set2',legend=False)
plt.title('Top Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales(INR)')
plt.show()
```

```python
#Sales contribution by channel
channel_sales=df.groupby('Channel ')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x=channel_sales.index,y=channel_sales.values,hue=channel_sales.index,palette='cubehelix',legend=False)
plt.title('Sales Contribution by Channel')
plt.xlabel('Channel')
plt.ylabel('Total Sales(INR)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Sales contribution by Category
category_sales=df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=category_sales.index,y=category_sales.values,hue=category_sales.index,palette='crest',legend=False)
plt.title('Highest Selling Category')
plt.xlabel('Category')
plt.ylabel('Total Sales(INR)')
plt.xticks(rotation=45)
plt.show()
```

```python
#Sales contribution by channel
b2b_b2c_sales=df.groupby('Business Type')['Amount'].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x='Business Type',y='Amount',data=b2b_b2c_sales,hue='Business Type',palette='Set2',legend=False)
plt.title('B2B vs B2C Transactions')
plt.xlabel('Business Type')
plt.ylabel('Total Sales(INR)')
plt.show()
```

This EDA confirms:The business is b2c heavy,women-centric,amazon-led,urban-focused,with Sets as the strongest product driver
